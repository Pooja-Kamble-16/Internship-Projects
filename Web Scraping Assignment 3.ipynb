# Question 2:
In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a dataframe and csv. In case if any product vertical has less than 3 pages in search results then scrape all the products available under that product vertical. Details to be scraped are: "Brand Name", "Name of the Product", "Rating", "No. of Ratings", "Price", "Return/Exchange", "Expected Delivery", "Availability", "Other Details" and “Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.

```python
url=[]
brand=[]
product=[]
ratings=[]
no_ratings=[]
price=[]
returns=[]
expected_delivery=[]
availability=[]
```

```python
for i in range(0,3): #Running for loop to scrap 3 pages
    
    #Using for loop to extract the url of each product with the xpath
    for j in driver.find_elements_by_xpath("//h2[@class='a-size-mini a-spacing-none a-color-base s-line-clamp-2']//a"):
        url.append(j.get_attribute('href'))
        
    driver.find_element_by_xpath("//li[@class='a-last']//a").click() #Clicking on the next button
    time.sleep(4)
```

```python
for i in url: #Loop for every URL in the list
    
    #Getting the webpage
    driver.get(i)
    
    try:         #exception handling for nosuchelementexception
        b=driver.find_element_by_xpath("//div[@class='a-expander-content a-expander-section-content a-section-expander-inner']//table[1]//tbody//tr[1]//td")  #Scraping the brand name with xpath
        brand.append(b.text)    #appending the brand in brand list
        
    except NoSuchElementException:
        brand.append("-")       #appending the "-" if no brand is found
        
    
    try:         #exception handling for nosuchelementexception
        p=driver.find_element_by_id("productTitle") #Scraping the product name with id
        product.append(p.text)  #appending the product name in product list
        
    except NoSuchElementException:
        product.append("-")     #appending the "-" if no product name is found
        
    
    try:         #exception handling for nosuchelementexception
        r=driver.find_element_by_xpath("//span[@class='a-size-base a-nowrap']//span")  #Scraping the ratings with xpath
        ratings.append(r.text)  #appending the ratings in ratings list
        
    except NoSuchElementException:
        ratings.append("-")     #appending the "-" if no rating is found
        
    
    try:         #exception handling for nosuchelementexception
        n=driver.find_element_by_xpath("//div[@class='a-row a-spacing-medium averageStarRatingNumerical']//span")  #Scraping the no_of_ratings with xpath
        no_ratings.append(n.text)  #appending the no of ratings in no_ratings list
    
    except NoSuchElementException:
        no_ratings.append("-")  #appending the "-" if no number of ratings is found
        
    
    try:         #exception handling for nosuchelementexception
        pr=driver.find_element_by_xpath("//span[@class='a-size-medium a-color-price priceBlockBuyingPriceString']")  #Scraping the price with xpath
        price.append(pr.text)   #appending the price in price list
    
    except NoSuchElementException:
        price.append("-")       #appending the "-" if price is found
        
    
    try:         #exception handling for nosuchelementexception
        re=driver.find_element_by_id("RETURNS_POLICY")
        returns.append(re.text) #appending the returns in returns list
    
    except NoSuchElementException:
        returns.append("-")    #appending the "-" if there is no returns/exchange
        
    
    try:         #exception handling for nosuchelementexception
        ed=driver.find_element_by_xpath("//div[@class='a-section a-spacing-mini']//b")  #Scraping expected delivery with xpath
        expected_delivery.append(ed.text)  #appending the expected delivery period in expected_delivery list
    
    except NoSuchElementException:
        expected_delivery.append("-") #appending the "-" if there is no expected delivery date mentioned
        

    try:         #exception handling for nosuchelementexception
        a=driver.find_element_by_id("availability") #Scraping the availability with id
        availability.append(a.text) #appending the availability in availability list

    except NoSuchElementException:
        availability.append("-") #appending the "-" if the product is not available
```

```python
#Combining all the lists in a DataFrame
amazon=pd.DataFrame({})
amazon["Brand"]=brand
amazon["Product Name"]=product
amazon["Price"]=price
amazon["Ratings"]=ratings
amazon["No of Ratings"]=no_ratings
amazon["Returns/Exchange"]=returns
amazon["Expected Delivery"]=expected_delivery
amazon["Availability"]=availability
amazon["Product URL"]=url
```

# Question 4:
Write a python program to search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) on www.flipkart.com and scrape following details for all the search results displayed on 1st page. Details to be scraped: “Brand Name”, “Smartphone name”, “Colour”, “RAM”, “Storage(ROM)”, “Primary Camera”, “Secondary Camera”, “Display Size”, “Display Resolution”, “Processor”, “Processor Cores”, “Battery Capacity”, “Price”, “Product URL”. Incase if any of the details is missing then replace it by “- “. Save your results in a dataframe and CSV.

```python
#Connecting the webdriver
driver=webdriver.Chrome("chromedriver.exe")
time.sleep(4)
```

```python
#Getting the required webpage
url ="https://www.flipkart.com/"
driver.get(url)
time.sleep(6)
```

```python
print("Enter smartphone")
phone=input() #Entering smartphone using input() saving it in a variable phone
```

```python
driver.find_element_by_xpath("//div[@class='_3OO5Xc']//input").send_keys(phone) #Sending the entered text as search element
```

```python
driver.find_element_by_xpath("//button[@class='L0Z3Pu']").click() #clicking the search button
```

```python
#Creating empty lists for the required data
urls=[]
brand=[]
smartphone=[]
colour=[]
RAM=[]
storage=[]
display_size=[]
display_res=[]
processor=[]
processor_core=[]
battery=[]
price=[]
camera=[]
```

```python
for i in driver.find_elements_by_xpath("//a[@class='_1fQZEK']"): #Running for loop to extract URL 
    urls.append(i.get_attribute('href'))
```

```python
for i in urls: #Loop for each URL in the list
    driver.get(i)
    
    driver.find_element_by_xpath("//button[@class='_2KpZ6l _1FH0tX']").click() #Clicking on the view more button
    
    try:         #exception handling for nosuchelementexception
        b=driver.find_element_by_xpath("//div[@class='_1UhVsV']//div[1]//table[1]//tbody//tr[3]//td[2]//ul//li") #Scraping brand name of the phone
        brand.append(b.text)     #appending brand name to the brand list
        
    except NoSuchElementException:
        brand.append("-")        #appending "-" if no brand name found
    
    try:         #exception handling for nosuchelementexception
        s=driver.find_element_by_xpath("//span[@class='B_NuCI']") #Scraping smartphone name
        smartphone.append(s.text.replace("/n"," "))  #appending smartphone name to the smartphone list
        
    except NoSuchElementException:
        smartphone.append("-")   #appending "-" if no smartphone name is found
        
    try:         #exception handling for nosuchelementexception
        col=driver.find_element_by_xpath("//div[@class='_1UhVsV']//div[1]//table[1]//tbody//tr[4]//td[2]//ul//li") #Scraping colour of the smartphone
        colour.append(col.text)  #appending colour of phone to the colour list
        
    except NoSuchElementException:
        colour.append("-")       #appending "-" if no colour is found
        
    try:         #exception handling for nosuchelementexception
        ram=driver.find_element_by_xpath("//div[@class='_1UhVsV']//div[4]//table[1]//tbody//tr[2]//td[2]//ul//li") #Scraping RAM of the smartphone
        RAM.append(ram.text)     #appending RAM to the RAM list
        
    except NoSuchElementException:
        RAM.append("-")          #appending "-" if RAM is not found
        
    try:         #exception handling for nosuchelementexception
        rom=driver.find_element_by_xpath("//div[@class='_1UhVsV']//div[4]//table[1]//tbody//tr[1]//td[2]//ul//li") #Scraping storage of the smartphone
        storage.append(rom.text) #appending storage to the storage list

    except NoSuchElementException:
        storage.append("-")      #appending "-" if no storage is found

    try:         #exception handling for nosuchelementexception
        cam=driver.find_element_by_xpath("//div[@class='_2418kt']//ul//li[3]") #Scraping camera details of the smartphone
        camera.append(cam.text)  #appending camera details to camera list
            
    except NoSuchElementException:
        camera.append("-")       #appending "-" if no camera details are found
    
    try:         #exception handling for nosuchelementexception
        dsize=driver.find_element_by_xpath("//div[@class='_1UhVsV']//div[2]//table[1]//tbody//tr[1]//td[2]//ul//li") #Scraping the Display Size
        display_size.append(dsize.text) #appending display size to the list
    
    except NoSuchElementException:
        display_size.append("-") #appending "-" if no display size is found
        
    try:         #exception handling for nosuchelementexception
        dres=driver.find_element_by_xpath("//div[@class='_1UhVsV']//div[2]//table[1]//tbody//tr[2]//td[2]//ul//li") #Scraping the display resolution
        display_res.append(dres.text) #appedning display resolution to the list

    except NoSuchElementException:
        display_res.append("-")  #appending "-" if no display resolution is found
        
    try:         #exception handling for nosuchelementexception
        pro=driver.find_element_by_xpath("//div[@class='_2418kt']//li[5]") #Scraping the smartphone processor
        processor.append(pro.text) #appending smartphone processor to the processor list

    except NoSuchElementException:
        processor.append("-")    #appending "-" if no processor is found
        
    try:         #exception handling for nosuchelementexception
        core=driver.find_element_by_xpath("//div[@class='_1UhVsV']//div[3]//table[1]//tbody//tr[3]//td[2]//ul//li") #Scraping the processor core
        processor_core.append(core.text) #appending processor core to the list

    except NoSuchElementException:
        processor_core.append("-") #appending "-" if no processor core is found
        
    try:         #exception handling for nosuchelementexception
        batt=driver.find_element_by_xpath("//div[@class='_2418kt']//li[4]") #Scraping the smartphone battery
        battery.append(batt.text) #appending smartphone battery to the battery list 

    except NoSuchElementException:
        battery.append("-")      #appedning "-" if no battery is found
        
    try:         #exception handling for nosuchelementexception
        p=driver.find_element_by_xpath("//div[@class='_30jeq3 _16Jk6d']") #Scraping the smartphone price
        price.append(p.text)     #appending price to the price list
    except NoSuchElementException:
        price.append("-")        #appending "-" if no price is found
```

```python
x=[]
#As Primary and Secondary camera details are captured together, splitting them and a list is created for each smartphone containing primary and secondary camera information.
for i in camera:
    x.append(i.split("|"))  #Storing the details in a list after split
```

```python
#Creating empty list for Primary camera and Secondary camera
pri_cam=[]
sec_cam=[]
```

```python
j=0  #initializing z as 0, so as to run loop from 0
for i in x: 
    pri_cam.append(x[j][0])   #appending first element of the list as primary camera info
    
    try:    #exception handling as some smartphones has no secondary camera
        sec_cam.append(x[j][1]) #appending second element of the list secondary camera info
    except:
        sec_cam.append("-")   #appending "-" if there is no secondary camera info
    j=j+1  #incremnting j by 1
```

```python
#Combining all the lists in a single Dataframe
phone=pd.DataFrame({})
phone["Brand Name"]=brand
phone["Smartphon Name"]=smartphone
phone["Colour"]=colour
phone["RAM"]=RAM
phone["Storage(ROM)"]=storage
phone["Primary Camera"]=pri_cam
phone["Secondary Camera"]=sec_cam
phone["Display Size"]=display_size
phone["Display Resolution"]=display_res
phone["Processor"]=processor
phone["Processor Core"]=processor_core
phone["Battery"]=battery
phone["Price"]=price
phone["Product URL"]=urls
```

Fetching the data using For loops

```python
#Scraping Product Name from the xpath
for i in driver.find_elements_by_xpath("//div[@class='right-container']//div//a//h3[1]"): 
    product.append(i.text)
    
#Scraping the Operating System of the laptop
for i in driver.find_elements_by_xpath("//div[@class='Spcs-details']//table[1]//tbody//tr[3]//td[3]"): 
    OS.append(i.text)
    
#Scraping display of the laptop
for i in driver.find_elements_by_xpath("//div[@class='Spcs-details']//table[1]//tbody//tr[4]//td[3]"):
    display.append(i.text)
    
#Scraping laptop processor
for i in driver.find_elements_by_xpath("//div[@class='Spcs-details']//table[1]//tbody//tr[5]//td[3]"):
    processor.append(i.text)
    
#Scraping the laptop Memory
for i in driver.find_elements_by_xpath("//div[@class='Spcs-details']//table[1]//tbody//tr[6]//td[3]"):
    memory.append(i.text)
    
#Scraping Weight of laptop
for i in driver.find_elements_by_xpath("//div[@class='Spcs-details']//table[1]//tbody//tr[7]//td[3]"):
    weight.append(i.text)
    
#Scraping laptop dimensions
for i in driver.find_elements_by_xpath("//div[@class='Spcs-details']//table[1]//tbody//tr[8]//td[3]"):
    dimension.append(i.text)
    
#Scraping Graphics Processor
for i in driver.find_elements_by_xpath("//div[@class='Spcs-details']//table[1]//tbody//tr[9]//td[3]"):
    graphics_processor.append(i.text)
    
#Scraping Price of Laptop
for i in driver.find_elements_by_xpath("//td[@class='smprice']"):
    price.append(i.text)
```

```python
#Creating dataframe for all the fetched data
laptops=pd.DataFrame({})
laptops["Product"]=product
laptops["Operating System"]=OS
laptops["Display"]=display
laptops["Processor"]=processor
laptops["Memory"]=memory
laptops["Weight"]=weight
laptops["Dimensions"]=dimension
laptops["Graphics Processor"]=graphics_processor
laptops["Price"]=price
```
