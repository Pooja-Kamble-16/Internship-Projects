# Python program to scrap book name, author name, genre and book review of any 5 books from ‘www.bookpage.com’

```python
page=requests.get("https://bookpage.com/reviews")
page
```

```python
soup=BeautifulSoup(page.content)
```

```python
book_name=[]
author=[]
genre=[]
book_review=[]
```

```python
review=soup.find_all('div',class_="bp-block article-info")
```

```python
for item in review:
    name=item.find('h4',class_="italic")
    auth=item.find('p',class_="sans bold")
    gen=item.find('p',class_="genre-links hidden-phone")
    rev=item.find('p',class_="excerpt")
    book_name.append(name.text.strip()) 
    author.append(auth.text.strip())
    genre.append(gen.text.replace('\n',' '))
    book_review.append(rev.text.replace('\n',' '))
```

```python
j=0
for i in book_review:
    if i==' ':
        book_name.remove(book_name[j])
        author.remove(author[j])
        genre.remove(genre[j])
        book_review.remove(book_review[j])
    j=j+1
```

```python
df=pd.DataFrame({})
df['Book Name']=book_name[0:5]
df['Author']=author[0:5]
df['Genre']=genre[0:5]
df['Reviews']=book_review[0:5]
```

```python
df
```

# Question :7

https://www.amazon.in/s?k=mobile+phones+under+20000&page=20&qid=1620645425&ref=sr_pg_20

```python
name=[]
price=[]
image=[]
rating=[]
```

```python
pages=list(range(1,21))
for page in pages:
    req=requests.get("https://www.amazon.in/s?k=mobile+phones+under+20000&qid=1620648182&ref=sr_pg_{}".format(page)).text
    soup=BeautifulSoup(req)
    
    prod=soup.find_all('span',class_="a-size-medium a-color-base a-text-normal")
    for i in range(len(prod)):
        name.append(prod[i].text.strip())
    
    pri=soup.find_all('span',class_="a-price-whole")
    for i in range(len(pri)):
        price.append(pri[i].text.strip())
        
    img=soup.find_all('img',class_="s-image")
    for i in range(len(img)):
        image.append(img[i])
    
    rate=soup.find_all('i',class_="a-icon a-icon-star-small a-star-small-4 aok-align-bottom")
    for i in range(len(rate)):
        rating.append(rate[i].text.strip())
```

```python
print(len(name))
print(len(price))
print(len(image))
print(len(rating))
```

```python
phones=pd.DataFrame({})
phones['Model']=name[0:91]
phones['Price']=price[0:91]
phones['Image']=image[0:91]
phones['Rating']=rating[0:91]
```

# Python program to extract information about the local weather from the National Weather Service website of USA, https://www.weather.gov/ for the city, San Francisco. You need to extract data about 7 day extended forecast display for the city. The data should include period, short description, temperature and description

```python
page=requests.get("https://forecast.weather.gov/MapClick.php?textField1=37.78&textField2=-122.42#.YJkLM7UzbIU")
page
```

```python
soup=BeautifulSoup(page.content)
```

```python
all_period=[]
all_short_desc=[]
all_temp=[]
all_desc=[]
```

```python
period=soup.find_all('p',class_="period-name")
short_desc=soup.find_all('p',class_="short-desc")
temp1=soup.find_all('p',class_="temp temp-low")
temp2=soup.find_all('p',class_="temp temp-high")
desc=soup.find_all('div',class_="col-sm-10 forecast-text")
```

```python
x=len(temp1+temp2)
j=0
for i in range(0,x):
    if i%2==0:
        all_temp.append(temp1[j].text.strip())
    else:
        all_temp.append(temp2[j].text.strip())
```

```python
for i in period:
    all_period.append(i.text.strip())
for i in short_desc:
    all_short_desc.append(i.text.strip())
for i in desc:
    all_desc.append(i.text.strip())
```

```python
fore=pd.DataFrame({})
fore['Period']=all_period
fore['Short Description']=all_short_desc
fore['Temperature']=all_temp
fore['Description']=all_desc[0:9]
```
