### 1. Scrape the details of most viewed videos on YouTube from Wikipedia: 
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos/

You need to find following details:

A) Rank
B) Name
C) Artist
D) Upload date
E) Views

```python
# Activating the chrome browser
driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(3)

# Opening the URL
url = "https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"
driver.get(url)
```

```python
# creating empty list for scraping data
rank=[]
name=[]
artist=[]
date=[]
views=[]
```

```python
#Running for loop to fetch data of the top 30 most viewed videos
for i in range(1,31):
    
    # Scraping the Rank 
    r=driver.find_element_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]//tbody//tr[{i}]//td[1]".format(i=i))
    rank.append(r.text.split(".")[0])
    
    # Scraping the Video Name
    n=driver.find_element_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]//tbody//tr[{i}]//td[2]".format(i=i))
    name.append(n.text)
    
    # Scraping the Artist
    a=driver.find_element_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]//tbody//tr[{i}]//td[3]".format(i=i))
    artist.append(a.text)
    
    # Scraping the Date 
    d=driver.find_element_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]//tbody//tr[{i}]//td[5]".format(i=i))
    date.append(d.text)
    
    # Scraping Views
    v=driver.find_element_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]//tbody//tr[{i}]//td[4]".format(i=i))
    views.append(v.text)
```

```python
# Creating dataframe for the scraped data of the most viewed YouTube videos
youtube=pd.DataFrame({})
youtube["Rank"]=rank
youtube["Name"]=name
youtube["Artist"]=artist
youtube["Upload Date"]=date
youtube["Views(Billion)"]=views

# Printing DataFrame
youtube
```

### 2. Scrape the details team India’s international fixtures from bcci.tv. 
Url = https://www.bcci.tv/. You need to find following details:

A) Match title (I.e. 1st ODI)
B) Series
C) Place
D) Date
E) Time

Note: - From bcci.tv home page you have reach to the international fixture page through code.


```python
# Activating the chrome browser
driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(3)

# Opening the URL
url = "https://www.bcci.tv/"
driver.get(url)
```

```python
#object of ActionChains
a=ActionChains(driver)
#Identifying element to hover
b=driver.find_element_by_xpath("//div[@class='navigation__menu-wrapper']/nav/ul/li[1]/div[1]")
#hover over element
a.move_to_element(b).perform()
driver.find_element_by_xpath("//div[@class='navigation__menu-wrapper']/nav/ul/li[1]/div[2]/div/ul/li[1]/a").click()
```

```python
# creating the empty lists for scraping data
title=[]
series=[]
place=[]
date=[]
time=[]
month=[]
```

```python
# Title
for i in driver.find_elements_by_xpath("//span[@class='u-unskewed-text fixture__tournament-label u-truncated']"):
    title.append(i.text)
    
# Series
for i in driver.find_elements_by_xpath("//strong[@class='fixture__name fixture__name--with-margin']"):
    series.append(i.text)
    
# Place
for i in driver.find_elements_by_xpath("//p[@class='fixture__additional-info']/span"):
    place.append(i.text)
    
# Date
for i in driver.find_elements_by_xpath("//span[@class='fixture__date']"):
    date.append(i.text.replace("\n"," "))
    
# Month
for i in driver.find_elements_by_xpath("//span[@class='fixture__month']"):
    month.append(i.text)
    
# Time
for i in driver.find_elements_by_xpath("//span[@class='fixture__time']"):
    time.append(i.text)
```

```python
date_mon=[]
for i in range(len(date)):
    date_mon.append(date[i]+" "+month[i]) # combining the Day and Month together to make Date
```

```python
# Creating dataframe for the scraped data of the India's International fixtures
India=pd.DataFrame({})
India["Match Title"]=title
India["Series"]=series
India["Place"]=place
India["Date"]=date_mon
India['Time']=time

# Printing DataFrame
India
```

### 4. Scrape the details of State-wise GDP of India from statisticstime.com.
Url = http://statisticstimes.com/
You have to find following details:

A) Rank
B) State
C) GSDP at current price (19-20)
D) GSDP at current price (18-19)

Note: - From statisticstimes home page you have to reach to economy page through code.

```python
# Activating the chrome browser
driver=webdriver.Chrome("chromedriver.exe")

# Opening the statisticstime.com url
url="http://statisticstimes.com/"
driver.get(url)
```

```python
# object of ActionChains
a = ActionChains(driver)

# Identifying element to hover
e=driver.find_element_by_xpath("//div[@class='navbar']/div[2]/button")

# hover over element
a.move_to_element(e).perform()
driver.find_element_by_xpath("//div[@class='navbar']/div[2]/div/a[3]").click()
```

```python
# clicking on "GDP of Indian States"
driver.find_element_by_xpath("//ul[@style='list-style-type:none;margin-left:20px;']/li[1]").click()
```

```python
# creating empty lists for scraping data
rank=[]
state=[]
GSDP_19_20=[]
GSDP_18_19=[]
share_18_19=[]
GDP=[]
```

```python
# Scraping Rank
for i in driver.find_elements_by_xpath("//tr[@class='odd' or @class='even']//td[1]"):
    rank.append(i.text)
    
# Scraping State
    for i in driver.find_elements_by_xpath("//tr[@class='odd' or @class='even']//td[2]"):
    state.append(i.text)

# Scraping GSDP at current price (19-20)
for i in driver.find_elements_by_xpath("//tr[@class='odd' or @class='even']//td[3]"):
    GSDP_19_20.append(i.text)
    
# Scraping GSDP at current price (18-19)
for i in driver.find_elements_by_xpath("//tr[@class='odd' or @class='even']//td[4]"):
    GSDP_18_19.append(i.text)

# Scraping Share(18-19)
for i in driver.find_elements_by_xpath("//tr[@class='odd' or @class='even']//td[5]"):
    share_18_19.append(i.text)

# Scraping GDP($ billion)
for i in driver.find_elements_by_xpath("//tr[@class='odd' or @class='even']//td[6]"):
    GDP.append(i.text)
```

```python
# creating dataframe for the scraped data of India's GDP
gdp=pd.DataFrame({})
gdp["Rank"]=rank[:33]
gdp["State"]=state[:33]
gdp["GSDP at current price (19-20)"]=GSDP_19_20[:33]
gdp["GSDP at current price (18-19)"]=GSDP_18_19[:33]
gdp["Share(18-19)"]=share_18_19[:33]
gdp["GDP(billion)"]=GDP[:33]

# printing dataframe
gdp
```

Note: - From the home page you have to click on the trending option from Explore menu through code.

```python
# Activating the chrome browser
driver=webdriver.Chrome("chromedriver.exe")

# Opening the Github URL
url="https://github.com/"
driver.get(url)
```

```python
# object of ActionChains
a=ActionChains(driver)

# Identifying element to hover
e=driver.find_element_by_xpath("/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/summary")

# hover over the element
a.move_to_element(e).click().perform()
driver.find_element_by_xpath("//ul[@class='d-lg-flex list-style-none']/li[4]/details/div/ul[2]/li[3]/a").click()
```

```python
urls=[]
title_1=[]
title_2=[]
desc=[]
contri=[]
lang=[]
list1=[]
```

```python
for i in driver.find_elements_by_xpath("//h1[@class='h3 lh-condensed']/a"):
    urls.append(i.get_attribute('href'))
    
for i in driver.find_elements_by_xpath("//span[@itemprop='programmingLanguage']"):
    lang.append(i.text)
```

```python
for url in urls:
    driver.get(url)
    
    t1=driver.find_element_by_xpath("//span[@class='author flex-self-stretch']/a")
    title_1.append(t1.text)
    
    t2=driver.find_element_by_xpath("//strong[@class='mr-2 flex-self-stretch']/a")
    title_2.append(t2.text)
        
    try:
        d=driver.find_element_by_xpath("//p[@class='f4 mt-3']")
        desc.append(d.text.replace("\n"," "))
    except NoSuchElementException:
        desc.append("-")
        
    try:
        if driver.find_element_by_xpath("//div[@class='BorderGrid BorderGrid--spacious']/div[6]/div/h2/a").text.split()[0]!="Contributors":
            if driver.find_element_by_xpath("//div[@class='BorderGrid BorderGrid--spacious']/div[5]/div/h2/a").text.split()[0]!="Contributors":
                if driver.find_element_by_xpath("//div[@class='BorderGrid BorderGrid--spacious']/div[4]/div/h2/a").text.split()[0]!="Contributors":
                    if driver.find_element_by_xpath("//div[@class='BorderGrid BorderGrid--spacious']/div[3]/div/h2/a").text.split()[0]=="Contributors":
                        c=driver.find_element_by_xpath("//div[@class='BorderGrid BorderGrid--spacious']/div[3]/div/h2/a/span")
                    else:
                        c=driver.find_element_by_xpath("//div[@class='BorderGrid BorderGrid--spacious']/div[2]/div/h2/a/span")
                else:
                    c=driver.find_element_by_xpath("//div[@class='BorderGrid BorderGrid--spacious']/div[4]/div/h2/a/span")
            else:
                c=driver.find_element_by_xpath("//div[@class='BorderGrid BorderGrid--spacious']/div[5]/div/h2/a/span")
        else:
            c=driver.find_element_by_xpath("//div[@class='BorderGrid BorderGrid--spacious']/div[6]/div/h2/a/span")
        contri.append(c.text)
    except NoSuchElementException:
        contri.append("-")
```

```python
title=[]
for i in range(len(title_1)):
    title.append(title_1[i]+"/"+title_2[i])
```

```python
# creating dataframe for the scraped data of repository
rep=pd.DataFrame({})
rep["Repository Title"]=title[:23]
rep["Respository Description"]=desc[:23]
rep["Contributors Count"]=contri[:23]
rep["Language Used"]=lang[:23]

# printing dataframe
rep
```

```python
# creating empty lists for scraping data
book_name=[]
author_name=[]
vol_sold=[]
publisher=[]
genre=[]
```

```python
for i in range(1,101):
    
    # Scraping Book name
    bn=driver.find_element_by_xpath("//table[@class='in-article sortable']//tbody//tr[{i}]//td[2]".format(i=i))
    book_name.append(bn.text)
    
    # Scraping Author name
    an=driver.find_element_by_xpath("//table[@class='in-article sortable']//tbody//tr[{i}]//td[3]".format(i=i))
    author_name.append(an.text)
    
    # Scraping Volumes sold
    vs=driver.find_element_by_xpath("//table[@class='in-article sortable']//tbody//tr[{i}]//td[4]".format(i=i))
    vol_sold.append(vs.text)
    
    # Scraping Publisher name
    p=driver.find_element_by_xpath("//table[@class='in-article sortable']//tbody//tr[{i}]//td[5]".format(i=i))
    publisher.append(p.text)
    
    # Scraping Genre
    gen=driver.find_element_by_xpath("//table[@class='in-article sortable']//tbody//tr[{i}]//td[6]".format(i=i))
    genre.append(gen.text)
```

### 10. Details of Datasets from UCI machine learning repositories.
Url = https://archive.ics.uci.edu/

You have to find the following details:

A) Dataset name
B) Data type
C) Task
D) Attribute type
E) No of instances
F) No of attribute
G) Year

Note: - from the home page you have to go to the Show All Dataset page through code.

```python
# Activating the chrome browser
driver=webdriver.Chrome("chromedriver.exe")

# OPening the URL
url="https://archive.ics.uci.edu/"
driver.get(url)
```

```python
# clicking on "view all the datasets"
driver.find_element_by_xpath("/html/body/table[2]/tbody/tr/td/span/b/a").click()
```

```python
# creating empty lists for scraping data
dataset=[]
datatype=[]
task=[]
attribute_type=[]
no_of_instances=[]
no_of_attribute=[]
year=[]
```

```python
for i in range(2,590):
    
    # Dataset
    ds=driver.find_element_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[{i}]/td[1]".format(i=i))
    dataset.append(ds.text)
    
    # Datatype
    dt=driver.find_element_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[{i}]/td[2]".format(i=i))
    datatype.append(dt.text)
    
    # Task
    t=driver.find_element_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[{i}]/td[3]".format(i=i))
    task.append(t.text)
    
    # Attribute Type
    a=driver.find_element_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[{i}]/td[4]".format(i=i))
    attribute_type.append(a.text)
    
    # No of Instances
    inst=driver.find_element_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[{i}]/td[5]".format(i=i))
    no_of_instances.append(inst.text)
    
    # No of Attribute
    na=driver.find_element_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[{i}]/td[6]".format(i=i))
    no_of_attribute.append(na.text)
        
    # Year
    y=driver.find_element_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[{i}]/td[7]".format(i=i))
    year.append(y.text)
```

```python
# Creating dataframe of the scraped data of datasets
UCI=pd.DataFrame({})
UCI["Dataset Name"]=dataset
UCI["Data type"]=datatype
UCI["Task"]=task
UCI["Attribute Type"]=attribute_type
UCI["No of Instances"]=no_of_instances
UCI["No of Attributes"]=no_of_attribute
UCI["Year"]=year

# printing dataset
UCI
```
